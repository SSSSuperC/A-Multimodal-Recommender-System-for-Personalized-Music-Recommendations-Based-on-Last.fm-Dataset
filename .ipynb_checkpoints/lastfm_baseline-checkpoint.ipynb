数据读取

```python
import pandas as pd

# Define file paths
data_path = "./Data"
artists_file = f"{data_path}/artists.dat"
tags_file = f"{data_path}/tags.dat"
user_artists_file = f"{data_path}/user_artists.dat"
user_taggedartists_file = f"{data_path}/user_taggedartists.dat"
user_friends_file = f"{data_path}/user_friends.dat"
user_taggedartists_timestamps_file = f"{data_path}/user_taggedartists-timestamps.dat"

# Load datasets with appropriate configurations
try:
    # Artists file
    artists = pd.read_csv(
        artists_file,
        sep='\t',
        header=0,  # Use the first row as column names
        encoding='utf-8'  # 保持 utf-8
    )

    # Tags file
    tags = pd.read_csv(
        tags_file,
        sep='\t',
        header=0,
        encoding='latin1',  # 使用宽容编码
        on_bad_lines='skip'  # 跳过结构异常行
    )

    # User-Artists file
    user_artists = pd.read_csv(
        user_artists_file,
        sep='\t',
        header=0,
        encoding='utf-8'
    )

    # User-Tagged Artists file
    user_taggedartists = pd.read_csv(
        user_taggedartists_file,
        sep='\t',
        header=0,
        encoding='utf-8'
    )

    # User-Friends file
    user_friends = pd.read_csv(
        user_friends_file,
        sep='\t',
        header=0,
        encoding='utf-8'
    )

    # User-Tagged Artists Timestamps file
    user_taggedartists_timestamps = pd.read_csv(
        user_taggedartists_timestamps_file,
        sep='\t',
        header=0,
        encoding='utf-8'
    )

    print("All datasets loaded successfully.")
except Exception as e:
    print(f"Error loading datasets: {e}")

# ===========================
# 提取独特的 userID
# ===========================

# 从各数据集中提取 userID
user_ids_from_artists = user_artists['userID'].unique()
user_ids_from_taggedartists = user_taggedartists_timestamps['userID'].unique()
user_ids_from_friends = user_friends['userID'].unique()

# 合并所有 userID 并去重
all_user_ids = pd.unique(
    pd.concat([
        pd.Series(user_ids_from_artists),
        pd.Series(user_ids_from_taggedartists),
        pd.Series(user_ids_from_friends)
    ])
)

# 创建用户 DataFrame
users = pd.DataFrame({'userID': all_user_ids})

# 保存用户表到文件
users.to_csv(f"{data_path}/user.csv", index=False)
print("\nUnique userID saved to 'users.csv'. Head of the file:")
print(users.head())
```

数据统计信息

```python
# Function to generate summary statistics for each dataset
def dataset_summary(dataset, dataset_name):
    # Generate missing values information
    missing_values = dataset.isnull().sum().to_dict()  # Number of missing values per column
    data_types = dataset.dtypes.to_dict()  # Data types of each column
    
    summary = {
        "Dataset": dataset_name,
        "Rows": dataset.shape[0],
        "Columns": dataset.shape[1],
        "Column Names": dataset.columns.tolist(),
        "Missing Values": missing_values,  # Ensure this is calculated
        "Data Types": data_types  # Ensure this is calculated
    }
    return summary

# Generate summaries for each dataset
datasets = {
    "artists": artists,
    "tags": tags,
    "user_artists": user_artists,
    "user_taggedartists": user_taggedartists,
    "user_friends": user_friends,
    "user_taggedartists_timestamps": user_taggedartists_timestamps
}

summaries = [dataset_summary(df, name) for name, df in datasets.items()]

# Display the summaries
for summary in summaries:
    print(f"Summary for {summary['Dataset']} Dataset:")
    print(f"Rows: {summary['Rows']}, Columns: {summary['Columns']}")
    print(f"Column Names: {summary['Column Names']}")
    print(f"Missing Values: {summary['Missing Values']}")
    print(f"Data Types: {summary['Data Types']}\n")
```

```python
import numpy as np

# ===========================
# Step 1: 缺失值处理
# ===========================
# 检查缺失值
missing_name = artists['name'].isnull().sum()
missing_pictureURL = artists['pictureURL'].isnull().sum()
print(f"\nMissing artist names: {missing_name}")
print(f"Missing artist picture URLs: {missing_pictureURL}")

# 处理 pictureURL 缺失值
artists['pictureURL'] = artists['pictureURL'].fillna("No Picture Available")

# ===========================
# Step 2: 检查重复值
# ===========================
# 确保 id 唯一性
duplicate_ids = artists['id'].duplicated().sum()
duplicate_names = artists['name'].duplicated().sum()
print(f"Duplicate artist IDs: {duplicate_ids}")
print(f"Duplicate artist names: {duplicate_names}")

# 去重（如果存在重复值）
artists = artists.drop_duplicates(subset='id')

# 将 id 列重命名为 artistID
artists.rename(columns={'id': 'artistID'}, inplace=True)

from sklearn.preprocessing import MinMaxScaler
import numpy as np

# ===========================
# Step 3: 提取文本特征并进行变换和标准化
# ===========================

# 统计名字中的单词数量
artists['word_count'] = artists['name'].str.split().apply(len)
# 统计名字的字符数量
artists['char_count'] = artists['name'].str.len()

# 检查 URL 是否有效（简单规则：非空 & 包含 "http"）
artists['pictureURL_valid'] = artists['pictureURL'].apply(lambda x: x != "No Picture Available").astype(int)

# 对 word_count 和 char_count 进行对数变换
artists['log_word_count'] = artists['word_count'].apply(lambda x: np.log1p(x))
artists['log_char_count'] = artists['char_count'].apply(lambda x: np.log1p(x))

# ===========================
# Step 4: 提取流行度特征并进行变换和标准化
# ===========================

# 特征 1: 艺术家被打标签的次数（标签流行度）
tag_counts = user_taggedartists_timestamps['artistID'].value_counts().reset_index()
tag_counts.columns = ['artistID', 'tag_count']

# 将 tag_counts 合并到 artists 数据集中
artists = artists.merge(tag_counts, on='artistID', how='left', suffixes=('', '_tag'))
artists['tag_count'] = artists['tag_count'].fillna(0)  # 对没有标签记录的艺术家填充为 0

# 特征 2: 艺术家的粉丝数（用户交互数）
fan_counts = user_artists['artistID'].value_counts().reset_index()
fan_counts.columns = ['artistID', 'fan_count']

# 将 fan_counts 合并到 artists 数据集中
artists = artists.merge(fan_counts, on='artistID', how='left', suffixes=('', '_fan'))
artists['fan_count'] = artists['fan_count'].fillna(0)  # 对没有用户交互的艺术家填充为 0

# 对 tag_count 和 fan_count 进行对数变换
artists['log_tag_count'] = artists['tag_count'].apply(lambda x: np.log1p(x))
artists['log_fan_count'] = artists['fan_count'].apply(lambda x: np.log1p(x))

# ===========================
# Step 5: 标准化所有数值特征
# ===========================

scaler = MinMaxScaler()

# 选择需要标准化的特征
features_to_normalize = ['log_word_count', 'log_char_count', 'log_tag_count', 'log_fan_count']

# 使用 MinMaxScaler 对这些特征进行标准化
artists[features_to_normalize] = scaler.fit_transform(artists[features_to_normalize])



# ===========================
# step 6:可视化分析
# ===========================

# 1. 可视化标签数量 (tag_count)
plt.figure(figsize=(10, 6))
plt.hist(artists['tag_count'], bins=50, alpha=0.75, color='blue')
plt.title('Distribution of Tag Counts per Artist')
plt.xlabel('Tag Count')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 对 tag_count 进行对数变换后的分布
plt.figure(figsize=(10, 6))
plt.hist(np.log1p(artists['tag_count']), bins=50, alpha=0.75, color='blue')
plt.title('Log-Transformed Distribution of Tag Counts per Artist')
plt.xlabel('Log(Tag Count)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. 可视化粉丝数量 (fan_count)
plt.figure(figsize=(10, 6))
plt.hist(artists['fan_count'], bins=50, alpha=0.75, color='green')
plt.title('Distribution of Fan Counts per Artist')
plt.xlabel('Fan Count')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 对 fan_count 进行对数变换后的分布
plt.figure(figsize=(10, 6))
plt.hist(np.log1p(artists['fan_count']), bins=50, alpha=0.75, color='green')
plt.title('Log-Transformed Distribution of Fan Counts per Artist')
plt.xlabel('Log(Fan Count)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3. 可视化对数变换后的粉丝数量 (log_fan_count)
plt.figure(figsize=(10, 6))
plt.hist(artists['log_fan_count'], bins=50, alpha=0.75, color='orange')
plt.title('Distribution of Log-Transformed Fan Counts per Artist')
plt.xlabel('Log(Fan Count)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 4. 可视化名字的单词数量 (word_count)
plt.figure(figsize=(10, 6))
plt.hist(artists['word_count'], bins=30, alpha=0.75, color='red')
plt.title('Distribution of Word Counts in Artist Names')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 对 word_count 进行对数变换后的分布
plt.figure(figsize=(10, 6))
plt.hist(np.log1p(artists['word_count']), bins=30, alpha=0.75, color='red')
plt.title('Log-Transformed Distribution of Word Counts in Artist Names')
plt.xlabel('Log(Word Count)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 5. 可视化名字的字符数量 (char_count)
plt.figure(figsize=(10, 6))
plt.hist(artists['char_count'], bins=30, alpha=0.75, color='purple')
plt.title('Distribution of Character Counts in Artist Names')
plt.xlabel('Character Count')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 对 char_count 进行对数变换后的分布
plt.figure(figsize=(10, 6))
plt.hist(np.log1p(artists['char_count']), bins=30, alpha=0.75, color='purple')
plt.title('Log-Transformed Distribution of Character Counts in Artist Names')
plt.xlabel('Log(Character Count)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# 6. 可视化图片有效性 (pictureURL_valid)
plt.figure(figsize=(8, 6))
plt.bar(['Valid Pictures', 'No Picture Available'], artists['pictureURL_valid'].value_counts(), color='green', alpha=0.85)
plt.title('Picture URL Validity Distribution')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



# ===========================
# Step 7: 保存预处理结果
# ===========================
# 保存处理后的数据集
processed_artists_features = artists[['artistID', 'tag_count', 'fan_count', 'log_fan_count', 'word_count', 'char_count', 
                                      'log_word_count', 'log_char_count', 'log_tag_count', 'pictureURL_valid']]

# 将处理后的特征保存到文件中
processed_artists_features.to_csv(f"{data_path}/processed_artists_features.csv", index=False)
print("\nProcessed Artists Features Dataset Saved!")
```

tag预处理

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: 清理未使用的标签
# 提取用户实际使用过的 tagID
used_tag_ids = user_taggedartists_timestamps['tagID'].unique()

# 筛选 tags 数据集，仅保留使用过的标签
filtered_tags = tags[tags['tagID'].isin(used_tag_ids)].copy()

# 更新 user_taggedartists-timestamps，确保与 filtered_tags 一致
filtered_user_taggedartists_timestamps = user_taggedartists_timestamps[
    user_taggedartists_timestamps['tagID'].isin(filtered_tags['tagID'])
].copy()

# Step 2: 合并标签数据
# 将标签值加入到交互数据中
merged_data = filtered_user_taggedartists_timestamps.merge(filtered_tags, on="tagID", how="left")

# Step 3: TF-IDF 特征提取
# 填充缺失值，避免空值问题
merged_data['tagValue'] = merged_data['tagValue'].fillna("")

# 构建每个用户的标签组合
user_tags = merged_data.groupby('userID')['tagValue'].apply(lambda x: ' '.join(x))

# 使用 TF-IDF 生成特征
tfidf_vectorizer = TfidfVectorizer(max_features=50)  # 限制最大特征数为 15
tfidf_features = tfidf_vectorizer.fit_transform(user_tags).toarray()

# 转换为 DataFrame 并添加 userID
tfidf_df = pd.DataFrame(tfidf_features, columns=[f'tfidf_{i}' for i in range(tfidf_features.shape[1])])
tfidf_df['userID'] = user_tags.index

# Step 4: 标签频率统计
# 统计每个标签的使用频率
tag_frequencies = merged_data['tagValue'].value_counts(normalize=True).to_dict()

# 将频率映射到每一行
merged_data['tag_frequency'] = merged_data['tagValue'].map(tag_frequencies)

# Step 5: 将特征整合到最终数据集
# 添加 TF-IDF 特征（基于 userID）
final_data = merged_data.merge(tfidf_df, on="userID", how="left")

# 查看结果
print("Final Data with TF-IDF and Tag Frequency:\n", final_data.head(20))

# Step 6: 可视化分析
def visualize_tag_features(final_data):
    # 标签频率分布
    plt.figure(figsize=(10, 6))
    final_data['tag_frequency'].plot(kind='hist', bins=80, color='teal', alpha=0.75)
    plt.title('Tag Frequency Distribution')
    plt.xlabel('Frequency')
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # 用户 TF-IDF 特征均值分布
    tfidf_columns = [col for col in final_data.columns if col.startswith('tfidf_')]
    tfidf_means = final_data[tfidf_columns].mean(axis=1)
    plt.figure(figsize=(10, 6))
    tfidf_means.plot(kind='hist', bins=50, color='orange', alpha=0.75)
    plt.title('TF-IDF Feature Mean Distribution')
    plt.xlabel('Mean TF-IDF Score')
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# 调用可视化
visualize_tag_features(final_data)

# ===========================
#  保存处理后的标签特征结果
# ===========================

# 选择包含用户 ID 和 TF-IDF 特征的列以及标签频率
processed_tag_features = final_data[['userID', 'tagValue', 'tag_frequency'] + [col for col in final_data.columns if col.startswith('tfidf_')]]

# 将处理后的特征保存到文件中
processed_tag_features.to_csv(f"{data_path}/processed_tag_features.csv", index=False)
print("\nProcessed Tag Features Dataset Saved!")
```

user_artists预处理

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define Data Path
data_path = "./Data"

# ===========================
# Step 1: 数据清理
# ===========================

# 检查缺失值
missing_values = user_artists.isnull().sum()
print("Missing values in user_artists:\n", missing_values)

# 检查重复值
duplicate_rows = user_artists.duplicated(subset=['userID', 'artistID']).sum()
print(f"Duplicate rows in user_artists: {duplicate_rows}")

# 如果存在重复值，删除重复行
if duplicate_rows > 0:
    user_artists = user_artists.drop_duplicates(subset=['userID', 'artistID'])

# ===========================
# Step 2: 特征工程（仅对数变换）
# ===========================

# 用户特征
user_features = user_artists.groupby('userID').agg(
    sum_weight=('weight', 'sum'),
    mean_weight=('weight', 'mean'),
    num_artists=('artistID', 'nunique')
).reset_index()

# 艺术家特征
artist_features = user_artists.groupby('artistID').agg(
    artist_popularity=('weight', 'sum'),
    num_users=('userID', 'nunique')
).reset_index()

# 对数变换
user_features['log_sum_weight'] = np.log1p(user_features['sum_weight'])
user_features['log_mean_weight'] = np.log1p(user_features['mean_weight'])
artist_features['log_artist_popularity'] = np.log1p(artist_features['artist_popularity'])
artist_features['log_num_users'] = np.log1p(artist_features['num_users'])

# ===========================
# Step 3: 保存处理后的数据
# ===========================

# 保存处理后的数据集
user_features.to_csv(f"{data_path}/processed_user_features.csv", index=False)
print("\nProcessed User Features Dataset Saved!")
artist_features.to_csv(f"{data_path}/processed_artist_features.csv", index=False)
print("\nProcessed Artist Features Dataset Saved!")

# ===========================
# Step 4: 可视化对数变换后的数据分布
# ===========================

def visualize_distribution(data, column, title):
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=50, alpha=0.75, color='skyblue')
    plt.title(f'Distribution of {title}')
    plt.xlabel(title)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# 可视化用户特征分布
visualize_distribution(user_features, 'log_sum_weight', 'Log Transformed Total Interaction Weight (User)')
visualize_distribution(user_features, 'log_mean_weight', 'Log Transformed Mean Interaction Weight (User)')

# 可视化艺术家特征分布
visualize_distribution(artist_features, 'log_artist_popularity', 'Log Transformed Artist Popularity')
visualize_distribution(artist_features, 'log_num_users', 'Log Transformed Number of Users (Artist)')
```

user_friends预处理

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define Data Path
data_path = "./Data"

# ===========================
# Step 1: 数据清理
# ===========================

# 检查缺失值
missing_values_friends = user_friends.isnull().sum()
print("Missing values in user_friends:\n", missing_values_friends)

# 检查重复值
duplicate_rows_friends = user_friends.duplicated(subset=['userID', 'friendID']).sum()
print(f"Duplicate rows in user_friends: {duplicate_rows_friends}")

# 如果存在重复值，删除重复行
if duplicate_rows_friends > 0:
    user_friends = user_friends.drop_duplicates(subset=['userID', 'friendID'])

# ===========================
# Step 2: 特征工程
# ===========================

# 用户朋友数量特征
user_friends_features = user_friends.groupby('userID').agg(
    num_friends=('friendID', 'nunique')
).reset_index()

# Step 2.1: 共享艺术家比例
# 建立用户与艺术家的映射
user_artist_map = user_artists.groupby('userID')['artistID'].apply(set).to_dict()

# 计算共享艺术家比例
def calculate_shared_artists_ratio(user_id, friends):
    if user_id not in user_artist_map:
        return 0.0
    user_artists = user_artist_map[user_id]
    shared_count = 0
    total_count = 0
    for friend_id in friends:
        if friend_id in user_artist_map:
            friend_artists = user_artist_map[friend_id]
            shared_count += len(user_artists & friend_artists)  # 交集
            total_count += len(friend_artists)
    return shared_count / total_count if total_count > 0 else 0.0

user_friends['shared_artists_ratio'] = user_friends.groupby('userID')['friendID'].transform(
    lambda friends: calculate_shared_artists_ratio(friends.name, friends)
)

# Step 2.2: 共享标签比例
# 建立用户与标签的映射（基于 user_taggedartists_timestamps 数据集）
merged_data = user_taggedartists_timestamps.merge(tags, on="tagID", how="left")
user_tag_map = merged_data.groupby('userID')['tagValue'].apply(set).to_dict()

# 计算共享标签比例
def calculate_shared_tags_ratio(user_id, friends):
    if user_id not in user_tag_map:
        return 0.0
    user_tags = user_tag_map[user_id]
    shared_count = 0
    total_count = 0
    for friend_id in friends:
        if friend_id in user_tag_map:
            friend_tags = user_tag_map[friend_id]
            shared_count += len(user_tags & friend_tags)  # 交集
            total_count += len(friend_tags)
    return shared_count / total_count if total_count > 0 else 0.0

user_friends['shared_tags_ratio'] = user_friends.groupby('userID')['friendID'].transform(
    lambda friends: calculate_shared_tags_ratio(friends.name, friends)
)

# Step 2.3: 聚合共享比例特征
shared_features = user_friends.groupby('userID').agg(
    mean_shared_artists_ratio=('shared_artists_ratio', 'mean'),
    mean_shared_tags_ratio=('shared_tags_ratio', 'mean')
).reset_index()

# 合并朋友数量和共享比例特征
user_friends_features = user_friends_features.merge(shared_features, on='userID', how='left')

# 对特征进行对数变换
user_friends_features['log_num_friends'] = np.log1p(user_friends_features['num_friends'])
user_friends_features['log_shared_artists_ratio'] = np.log1p(user_friends_features['mean_shared_artists_ratio'])
user_friends_features['log_shared_tags_ratio'] = np.log1p(user_friends_features['mean_shared_tags_ratio'])

# ===========================
# Step 3: 保存处理后的数据
# ===========================

# 保存用户朋友特征
user_friends_features.to_csv(f"{data_path}/processed_user_friends.csv", index=False)
print("\nProcessed User Friends Dataset Saved!")

# ===========================
# Step 4: 可视化特征分布
# ===========================

def visualize_user_friends_features(data, column, title):
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=50, alpha=0.75, color='purple')
    plt.title(f'Distribution of {title}')
    plt.xlabel(title)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# 可视化朋友数量分布
visualize_user_friends_features(user_friends_features, 'log_num_friends', 'Log Transformed Number of Friends (User)')
# 可视化共享艺术家比例分布
visualize_user_friends_features(user_friends_features, 'log_shared_artists_ratio', 'Log Transformed Shared Artists Ratio (User)')
# 可视化共享标签比例分布
visualize_user_friends_features(user_friends_features, 'log_shared_tags_ratio', 'Log Transformed Shared Tags Ratio (User)')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

# ===========================
# Step 1: 构建艺术家-标签关系
# ===========================

# 合并 user_taggedartists_timestamps 和 tags 数据集
artist_tag_data = user_taggedartists_timestamps.merge(tags, on="tagID", how="left")

# 统计每个艺术家被打的标签及其频率
artist_tag_counts = artist_tag_data.groupby(['artistID', 'tagValue']).size().reset_index(name='tag_count')

# 计算标签的全局权重（所有艺术家中的频率）
global_tag_weights = artist_tag_counts.groupby('tagValue')['tag_count'].sum().reset_index()
global_tag_weights.rename(columns={'tag_count': 'global_tag_count'}, inplace=True)

# 将全局权重归一化
global_tag_weights['global_tag_weight'] = global_tag_weights['global_tag_count'] / global_tag_weights['global_tag_count'].sum()

# 合并全局权重到艺术家-标签数据
artist_tag_counts = artist_tag_counts.merge(global_tag_weights, on='tagValue', how='left')

# 归一化每个艺术家的标签频率
artist_tag_counts['normalized_tag_weight'] = artist_tag_counts['tag_count'] / artist_tag_counts.groupby('artistID')['tag_count'].transform('sum')

# 查看结果
print("Artist-Tag Counts (with Normalized Weights):\n", artist_tag_counts.head())

# ===========================
# Step 2: 标签特征增强（TF-IDF）
# ===========================

# 构建艺术家和用户的标签文本
artist_tags_text = artist_tag_data.groupby('artistID')['tagValue'].apply(lambda x: ' '.join(x)).reset_index()
user_tags_text = artist_tag_data.groupby('userID')['tagValue'].apply(lambda x: ' '.join(x)).reset_index()

# 使用统一的 TF-IDF 提取特征
tfidf_vectorizer = TfidfVectorizer(max_features=50)  # 限制最大特征数

# 为艺术家生成标签特征
artist_tfidf_features = tfidf_vectorizer.fit_transform(artist_tags_text['tagValue']).toarray()
artist_tag_matrix = pd.DataFrame(artist_tfidf_features, columns=[f"tag_{i}" for i in range(artist_tfidf_features.shape[1])])
artist_tag_matrix['artistID'] = artist_tags_text['artistID']

# 为用户生成标签特征
user_tfidf_features = tfidf_vectorizer.transform(user_tags_text['tagValue']).toarray()
user_tag_matrix = pd.DataFrame(user_tfidf_features, columns=[f"tag_{i}" for i in range(user_tfidf_features.shape[1])])
user_tag_matrix['userID'] = user_tags_text['userID']

# ===========================
# Step 3: 用户-标签关系增强
# ===========================

# 合并用户与艺术家交互数据
user_artist_tags = user_artists.merge(artist_tag_counts, on='artistID', how='left')

# 用户标签偏好：根据用户对艺术家的权重计算加权平均
user_tag_preferences = user_artist_tags.groupby(['userID', 'tagValue']).apply(
    lambda x: np.average(x['normalized_tag_weight'], weights=x['weight'])
).reset_index(name='user_tag_preference')

# 为用户生成标签偏好的特征向量
user_preference_matrix = user_tag_preferences.pivot(index='userID', columns='tagValue', values='user_tag_preference').fillna(0)

# ===========================
# Step 4: 保存处理后的数据
# ===========================

# 保存用户标签矩阵
user_tag_matrix.to_csv(f"{data_path}/processed_user_tag_matrix.csv", index=False)
print("\nProcessed User-Tag Matrix Saved!")

# 保存艺术家标签矩阵
artist_tag_matrix.to_csv(f"{data_path}/processed_artist_tag_matrix.csv", index=False)
print("\nProcessed Artist-Tag Matrix Saved!")

# 保存用户偏好矩阵
user_preference_matrix.to_csv(f"{data_path}/processed_user_preference_matrix.csv", index=True)
print("\nProcessed User Preference Matrix Saved!")

# ===========================
# Step 5: 可视化
# ===========================

def visualize_matrix(matrix, title, index_sample=50, column_sample=50):
    plt.figure(figsize=(12, 8))
    plt.imshow(matrix.iloc[:index_sample, :column_sample].values, aspect='auto', cmap='coolwarm')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('Tags')
    plt.ylabel('Users/Artists')
    plt.show()

# 可视化前 50x50 的用户-标签矩阵
visualize_matrix(user_tag_matrix.set_index('userID'), "User-Tag Matrix (Sample)")

# 可视化前 50x50 的艺术家-标签矩阵
visualize_matrix(artist_tag_matrix.set_index('artistID'), "Artist-Tag Matrix (Sample)")
```

MLP模型

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

class TransformerRecommender(nn.Module):
    def __init__(self, input_dim, num_classes, nhead=8, num_encoder_layers=4, dim_feedforward=128, dropout=0.1):
        super(TransformerRecommender, self).__init__()
        self.input_layer = nn.Linear(input_dim, dim_feedforward)
        
        # 使用多个 Transformer 编码器层
        encoder_layer = nn.TransformerEncoderLayer(d_model=dim_feedforward, nhead=nhead, dim_feedforward=dim_feedforward, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_encoder_layers)
        
        # 输出类别数（多分类）
        self.fc = nn.Linear(dim_feedforward, num_classes)

    def forward(self, x):
        # 初始线性变换
        x = self.input_layer(x).unsqueeze(1)  # 增加一个维度作为 sequence length
        # Transformer 编码器层
        x = self.transformer_encoder(x).squeeze(1)  # 移除 sequence length 维度
        # 最后通过全连接层
        return self.fc(x)
```

MLP模型训练

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from torch.utils.data import DataLoader, TensorDataset

# 重新加载最终特征表
data_path = "./Data"
interaction_data = pd.read_csv(f"{data_path}/final_interaction_data.csv")

# 检查是否存在 NaN 或 inf 值
interaction_data = interaction_data.fillna(0)
interaction_data.replace([np.inf, -np.inf], 0, inplace=True)

# 特征列和目标列
feature_columns = interaction_data.columns.difference(['userID', 'artistID'])
X = interaction_data[feature_columns].values
y = interaction_data['artistID'].values

# 保存 userID
user_ids = interaction_data['userID'].values

# 使用 LabelEncoder 将 artistID 重新编码为从 0 开始的整数索引
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 数据标准化
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 获取类别数
num_classes = len(label_encoder.classes_)

# 划分训练集、验证集和测试集，并保存测试集的 userID
X_train, X_temp, y_train, y_temp, user_ids_train, user_ids_temp = train_test_split(X, y_encoded, user_ids, test_size=0.3, random_state=42)
X_val, X_test, y_val, y_test, user_ids_val, user_ids_test = train_test_split(X_temp, y_temp, user_ids_temp, test_size=0.5, random_state=42)

# 数据转换为 PyTorch 格式
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)  # 多分类任务需要 long 类型标签
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)

# 定义模型、损失函数和优化器
transformer_model = TransformerRecommender(input_dim=X_train.shape[1], num_classes=num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(transformer_model.parameters(), lr=0.001)

# 记录损失和准确率
train_losses, val_losses = [], []
train_recalls, val_recalls = [], []

# 模型训练
epochs = 50
for epoch in range(epochs):
    transformer_model.train()
    epoch_loss = 0
    correct_train, total_train = 0, 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        predictions = transformer_model(X_batch)
        loss = criterion(predictions, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

        # 计算训练集 Recall
        _, predicted = torch.max(predictions, 1)
        correct_train += (predicted == y_batch).sum().item()
        total_train += y_batch.size(0)

    train_loss = epoch_loss / len(train_loader)
    train_recall = correct_train / total_train
    train_losses.append(train_loss)
    train_recalls.append(train_recall)

    # 验证集损失和 Recall
    transformer_model.eval()
    with torch.no_grad():
        val_predictions = transformer_model(X_val_tensor)
        val_loss = criterion(val_predictions, y_val_tensor)
        _, val_predicted = torch.max(val_predictions, 1)
        correct_val = (val_predicted == y_val_tensor).sum().item()
        val_recall = correct_val / y_val_tensor.size(0)
    
    val_losses.append(val_loss.item())
    val_recalls.append(val_recall)

    print(f"Epoch {epoch+1}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Train Recall: {train_recall:.4f}, Val Recall: {val_recall:.4f}")

# 测试集预测
transformer_model.eval()
with torch.no_grad():
    y_pred_transformer = transformer_model(X_test_tensor)
    y_pred_transformer = torch.argmax(y_pred_transformer, axis=1)
    correct_test = (y_pred_transformer == y_test_tensor).sum().item()
    accuracy = correct_test / y_test_tensor.size(0)
    print(f"Test Accuracy: {accuracy:.4f}")

# 将预测的 artistID 重新映射回原始的 artistID
y_pred_artistID = label_encoder.inverse_transform(y_pred_transformer.numpy())

# 保存预测结果
interaction_data_test = pd.DataFrame(X_test, columns=feature_columns)
interaction_data_test['userID'] = user_ids_test
interaction_data_test['predicted_artistID'] = y_pred_artistID

interaction_data_test.to_csv(f"{data_path}/user_artist_recommendations_transformer.csv", index=False)
print("Recommendations saved to user_artist_recommendations_transformer.csv")

# 可视化训练过程
plt.figure(figsize=(12, 5))

# 绘制 Loss 曲线
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), train_losses, label='Train Loss')
plt.plot(range(1, epochs + 1), val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# 绘制 Recall 曲线
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), train_recalls, label='Train Recall')
plt.plot(range(1, epochs + 1), val_recalls, label='Val Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.title('Training and Validation Recall')
plt.legend()

plt.tight_layout()
plt.show()
```
